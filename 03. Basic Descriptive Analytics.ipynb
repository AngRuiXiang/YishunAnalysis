### Comparison of Numbers by Published Year

**All Articles**

```python
year_count_all = df['Published Year'].value_counts().sort_index()

fig, ax = plt.subplots(figsize=(18,10))
bar_container = ax.bar(year_count_all.index, year_count_all.values);

ax.set_title('Number of Singapore News Articles by Year')
ax.set_xlabel('Year')
ax.set_ylabel('Number of News Articles')
ax.bar_label(bar_container)

plt.xticks(df['Published Year'].unique());
```

**Articles involving Yishun Only**

```python
year_count_yishun = df[df['Town Involved']=='Yishun']['Published Year'].value_counts().sort_index()

fig, ax = plt.subplots(figsize=(18,10))
bar_container = ax.bar(year_count_yishun.index, year_count_yishun.values);

ax.set_title('Number of Singapore News Articles Involving Yishun by Year')
ax.set_xlabel('Year')
ax.set_ylabel('Number of News Articles')
ax.bar_label(bar_container)

plt.xticks(df['Published Year'].unique());
```

**Articles in Courts-crime Category Only**

```python
year_count_courts = df[df['Category']=='courts-crime']['Published Year'].value_counts().sort_index()

fig, ax = plt.subplots(figsize=(18,10))
bar_container = ax.bar(year_count_courts.index, year_count_courts.values);

ax.set_title('Number of Singapore News Articles in Courts-Crime Category by Year')
ax.set_xlabel('Year')
ax.set_ylabel('Number of News Articles')
ax.bar_label(bar_container)

plt.xticks(df['Published Year'].unique());
```

### Comparison of Numbers by Town

**Absolute Numbers**

```python
town_count_all = df['Town Involved'].value_counts().sort_index()

fig, ax = plt.subplots(figsize=(18,10))
bar_container = ax.bar(town_count_all.index, town_count_all.values);

ax.set_title('Number of Singapore News Articles by Town Involved')
ax.set_xlabel('Town')
ax.set_ylabel('Number of News Articles')
ax.bar_label(bar_container)

plt.xticks(rotation=90);
```

Comment: While Yishun do have quite a high number of news article, it is not the highest. It would make better sense to average out these numbers by population - after all, it's natural for places with higher population to have more news. However, a caveat is that since the news are tracked by HDB town, the population by HDB town naturally only reflects those people who stay in HDB flats. This means that towns with more private property residents would not be accurately counted.

Data on population per HDB town can be obtained from data.gov: https://beta.data.gov.sg/datasets/d_0a6c6d71f6fa14e2d27e406f1d018439/view. Unfortunately, it was only updated until 2018

**Number of Articles per 100,000 population**

```python
hdb_population = pd.read_csv('HDBResidentPopulationbyGeographicalDistribution.csv')
hdb_population.head()
```

```python
# Latest year of the population data

max(hdb_population['shs_year'])
```

```python
hdb_population = hdb_population[hdb_population['shs_year'] == max(hdb_population['shs_year'])]
hdb_population
```

```python
population_count = {}

for town in hdb_population['town_estate'].unique():
    population_count[town.lower().replace('_', ' ')] = hdb_population[hdb_population['town_estate'] == town]['number'].values[0]
    
population_count['kallang'] = population_count.pop('kallang whampoa')
population_count
```

```python
# Towns not covered under population_count

list(set(town_count_all.index.str.lower()) - set(population_count.keys()))
```

```python
articles_per_population = {}

for i in town_count_all.index:
    if i.lower() in population_count.keys():
        articles_per_population[i] = round((town_count_all[i] / population_count[i.lower()]) * 100000)
    
articles_per_population
```

```python
fig, ax = plt.subplots(figsize=(18,10))
bar_container = ax.bar(articles_per_population.keys(), articles_per_population.values());

ax.set_title('Number of Singapore News Articles per 100,000 HDB Residents by Town Involved')
ax.set_xlabel('Town')
ax.set_ylabel('Number of News Articles')
ax.bar_label(bar_container)

plt.xticks(rotation=90);
```

Comment: Bukit Timah likely has the highest number of news article per 100,000 population because the data are wrong - the population only covers HDB residents and Bukit Timah has a lot of private property residences that are not captured. This chart should thus be taken with a pinch of salt. I am not able to find data that covers non-HDB population by HDB town.

Ignoring Bukit Timah, we see that Yishun still has high number of articles, but is not the highest
